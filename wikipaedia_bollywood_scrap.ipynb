## we will try to scrap the wikipaedia page containing the list of bollywood films

```python
import requests
page = requests.get("https://en.wikipedia.org/wiki/Lists_of_Bollywood_films")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for url in soup.find_all('a'):
    print(url.get('href'))
```

## Now we will try to find the number of films produced in the last 100 years in bollywood.

```python
import re
cnt= 0
film_list = []
for url in soup.findAll('a', href = re.compile("List_of_Bollywood_films")):
    actual_link = "https://en.wikipedia.org" + url.get('href')
    print(actual_link)
    cnt = cnt + 1
    film_list.append(actual_link)
    if cnt>100:
        break
        
```

```python
import requests
page_tab = requests.get("https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_1989")

soup_tab = BeautifulSoup(page_tab.content, 'html.parser')
```

```python
tables = soup_tab.find_all('table')
```

```python
print(tables[2].prettify())
```

```python
total = 0
for film in film_list:
    page_tab = requests.get(film)

    soup_tab = BeautifulSoup(page_tab.content, 'html.parser')
    tables = soup_tab.find_all('table')
    lis = tables[2].find_all('tr')
    num = len(lis)
    total = total + num
    
print(total)    
```
