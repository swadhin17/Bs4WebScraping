## scraping the second page

```python
import requests
page = requests.get("https://www.flipkart.com/mens-footwear/casual-shoes/pr?sid=osp%2Ccil%2Ce1f&otracker=nmenu_sub_Men_0_Casual+Shoes&page=2")

from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
divs = soup.find_all('div' , {'class':'_3O0U0u'})
```

```python
l = len(divs)
print(l)
```

```python
for i in range(0,l):
    comps = divs[i].find_all('div' , {'class':'_2B_pmu'})
    names = divs[i].find_all('a' , {'class':'_2mylT6'})
    prices = divs[i].find_all('div' , {'class':'_1vC4OE'})
    k = len(comps)
    for j in range(0,k):
        c = comps[j].text
        n = names[j].text
        p = prices[j].text
        print(c, n,'\t', p)
```

### The above list shows the name of the company, model and prices of varoius shoes in flipkart.com website.
