## We will scrap a page from imdb official website

#### Refer to webpage https://www.imdb.com/list/ls064417595/

```python
import requests
page = requests.get("https://www.imdb.com/list/ls064417595/")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
divs = soup.find_all('div' , {'class':'lister-item mode-detail'})
```

```python
k = len(divs)
```

```python
divs
```

```python
import re
def get_name(i):
    names = divs[i].find_all('a' , href = re.compile('/title/'))
    name = names[1].text
    return name
```

```python
def get_ratings(i):
    ratings = divs[i].find_all("span", {"class": "ipl-rating-star__rating"})
    rate = ratings[0].text
    return rate
```

```python
for n in range(0,k):
    name_given = get_name(n)
    rate_given = get_ratings(n)
    print(name_given , rate_given)
    
```
